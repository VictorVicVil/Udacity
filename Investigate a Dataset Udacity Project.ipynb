We'll have to drop those rows before merging the data sets 

```python
def reduce_df(df):
    rows=df.index.tolist()
    for i in rows:
        i=0
        if i<=len(rows):
            reduced_df=suicidemen_df.drop(suicidemen_df.index[rows[i]])
            i=i+1
        return reduced_df
```
